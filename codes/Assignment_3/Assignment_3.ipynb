```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def my_exp(x, n=100):
    sum_n = 1
    tolerance = 1e-14
    for i in range(1,n+1):
        sum_p = sum_n
        sum_n = sum_n+(x**i/math.factorial(i))
        delta = sum_n - sum_p
        if delta<tolerance:
            break
        
    return  sum_n
```

```python
my_exp(15)
```

```python
x = np.linspace(0,100,1000)
y1= np.exp(x)
y2 = np.zeros((1000))
for i in range(1000):
    y2[i] = my_exp(x[i])

```

```python
plt.figure(figsize=(12,7))
plt.plot(x,y1,label="numpy exp() func",markerfacecolor ="Red")
plt.plot(x,y2,label="my_exp() func",markerfacecolor ="Black")
plt.xlabel("x")
plt.ylabel("exp(x)")
plt.legend()
plt.show()
```
